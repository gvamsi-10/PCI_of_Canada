## data visualization

```python
%matplotlib inline
plt.xlabel('pca')
plt.ylabel('year')
plt.scatter(df.year,df.pca,color='red',marker='*')
plt.plot(df.year,df.pca,color='green')
```

```python
y = df.pca
y
```

```python
X = pd.DataFrame(df.year)  # np.array(df.drop('pca', axis=1)), df.drop('pca', axis=1)
X
```

## method-1

```python
res = linear_model.LinearRegression()
res.fit(X,y)
```

```python
res.coef_
```

```python
res.intercept_
```

```python
res.score(X,y)
```

### new_predictions

```python
res.predict([[2017],[2018],[2019],[2020],[2021],[2022]])
```
